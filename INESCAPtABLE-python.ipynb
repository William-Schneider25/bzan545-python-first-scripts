## An Aside for Encryption Tools

You'll need some tools before you start investigating.

Run the below cell to load some decryption tools!

````python
ALPHABET = (
    list("abcdefghijklmnopqrstuvwxyz")
    + list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
    + list("123456789")
    + [" ", ".", "'", '"', "?", "&", "$", "#", "!", ","]
)


def normalize_index(i, n):
    return i % n


def encode(message, key):
    if pd.isna(message) or pd.isna(key):
        return None
    key = int(key)
    chars = []
    for char in str(message):
        if char not in ALPHABET:
            chars.append(char)
            continue
        old_i = ALPHABET.index(char)
        new_i = normalize_index(old_i + key, len(ALPHABET))
        chars.append(ALPHABET[new_i])
    return "".join(chars)


def decode(message, key):
    return encode(message, -int(key))


req_funcs = ["normalize_index", "encode", "decode"]
assert all(func in globals() for func in req_funcs)
````

The encrypted messages you'll encounter were encoded using a "Caesar cipher".  A Caesar cipher is a simple way to encrypt info by shifting the alphabet.  If you want to learn more about it check out this short video: https://www.youtube.com/watch?v=l6jqKRXSShI

We don't have to concern ourselves with that too much though.  Here's how we'll use the functions loaded for us above.

````python
# Encrypt a message
encoded = encode("abc", key=2)
print(encoded)

# Decrypt the encrypted message using the same key
decode(encoded, key=2)
````

````python
msg = "Rocky Top, you'll always be home sweet home to me!!!"

# Encrypt a message
encoded = encode(msg, key=865)
print(encoded)

# Decrypt the encrypted message using the same key
decode(encoded, key=865)
````

## Gameplay Examples: Path 1

The basic loop is:

1. Find the row for the item.
2. Figure out the key.
3. Decode the message.
4. Use that message to find the next item and key.

We'll walk through a few examples before you take over.

### Current item: Refrigerator

````python
row = manor.loc[(manor["room"] == "Kitchen") & manor["key"].notna()]
key = row["key"].iloc[0]

decode(row["inspect"].iloc[0], key)
````

^ This points to our next item being `"Piano"` and its key being `101`.

### Current item: Piano

To find the piano, we could list all items and look for it, or guess that the item name is `"Piano"` based on the clue wording/casing.

````python
manor["item"].values
````

````python
row = manor.loc[manor["item"] == "Piano"]
key = 101

decode(row["inspect"].iloc[0], key)
````

This clue gives us both a next item and a key calculation. The next item is `"Bar Cart"`; the key is the square root of `12321` times the number of piano keys.

### Current item: Bar Cart

A standard piano has 88 keys (yeah, i put trivia in here to - you can search it online ofc), so the clue becomes `sqrt(12321) * 88`.

````python
row = manor.loc[manor["item"] == "Bar Cart"]
key = (12321**0.5) * 88

decode(row["inspect"].iloc[0], key)
````

### Current item: Fireplace

Now the key is not written directly. We have to create it from the `weight_kg` column.

Use a copy of the weight column so we can replace missing values without changing `manor` itself.

````python
row = manor.loc[manor["item"] == "Fireplace"]

wts = manor["weight_kg"].copy()
wts = wts.fillna(wts.median())
key = round(wts.mean())

decode(row["inspect"].iloc[0], key)
````

The decoded clue sends us to the Ladder. This one is a math clue, not a pandas clue.

### Current item: Ladder

The clue describes a right triangle: 7 feet from the wall and 24 feet up the wall.

````python
row = manor.loc[manor["item"] == "Ladder"]
key = (7**2 + 24**2) ** 0.5

decode(row["inspect"].iloc[0], key)
````
